```python
#Menampilkan Film yang Dirental Mary Smith dengan Payment Amount lebih besar dari 4.0
dataFilm = db.Sakila_customers.find({"First Name" : "ANA", "Last Name" : "BRADLEY"}, {'Rentals.Payments.Amount', "Rentals.Film Title"})
RentalLen = len(dataFilm[0]['Rentals'])
minimum = 4.0
amounts = []
paymentGreatherThan4 = []
for l in range(RentalLen):
    amount = dataFilm[0]['Rentals'][l]['Payments'][0]['Amount']
    if float(amount) > minimum :
        amounts.append(amount)
        paymentGreatherThan4.append(dataFilm[0]['Rentals'][l])


amounts.sort()
excludedTitle = []
dataSortedByPayments = []
for i in range(len(amounts)):
    for j in range(len(paymentGreatherThan4)) :
        if amounts[i] == paymentGreatherThan4[j]["Payments"][0]["Amount"] :
            if (paymentGreatherThan4[j]['Film Title'] in excludedTitle) == False:
                dataSortedByPayments.append(paymentGreatherThan4[j])
                excludedTitle.append(paymentGreatherThan4[j]['Film Title'])

DataFrame(dataSortedByPayments)
```
